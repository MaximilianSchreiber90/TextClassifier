```python
# all relevant functions are in access_data.py
# raw datasets:
# - get_20news() for all datasets from 20 newsgroup dataset
# - huff_prepared() for all datasets from huffpost
# cleaning:
# - clean_text(df) using WordNetLemmatizer and remove non letters
# - add_word_vector_column(df)  calls clean_text() and puts result into new ['words'] column
# - delete_short_datasets(df, minWords)  removes datasets with small (>int) words
# all in one function:
# - get_all_data_prepared(minWords)  returns all datasets (for multilabel classification)
# - get_df_map(ratio) returns a map<labelname, df> of binary labels and their dataset

datasets = get_df_map()
```

# Getting best classifier

```python
from sklearn.svm import LinearSVC

bestClassifier_svm_config = do_gridsearch(datasets, LinearSVC())
bestClassifier_logReg_config = do_gridsearch(datasets, LogisticRegression())
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from matplotlib import pyplot

from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_curve
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC

classifierMap = {}

for key, df in datasets.items():
    #splitting data
    split_map_all = split_dataset(df)
    X_train_all = split_map_all['X_train']
    y_train_all = split_map_all['y_train']
    X_test_all = split_map_all['X_test']
    y_test_all = split_map_all['y_test']

    #encoding label
    integer_encoded_train = y_train_all.replace("NONE", 0).replace(key, 1)
    integer_encoded_test = y_test_all.replace("NONE", 0).replace(key, 1)
    
    tfidf_vectorizer =  TfidfVectorizer(stop_words='english', max_df=0.15, min_df=5)
    X_train = tfidf_vectorizer.fit_transform(X_train_all)
    X_test = tfidf_vectorizer.transform(X_test_all)
    
    from sklearn.decomposition import TruncatedSVD
    svd = TruncatedSVD(n_components=2500, n_iter=4, random_state=42)    # there are 52120 features
#    X_train_SVD = svd.fit_transform(X_train)
#    X_test_SVD = svd.transform(X_test)

    svc_f1 = 0
    if bestClassifier_svm_config is None:
        print('No LinearSVC classifier found for label {}'.format(key))
    else:
        svc = bestClassifier_svm_config[key].svc
        svc.fit(X_train, integer_encoded_train)
        probs_svc = svc.predict(X_test)

        print_AUC(integer_encoded_test, probs_svc, key)
        print_prec_rec(integer_encoded_test, probs_svc)
        svc_f1 = f1_score(integer_encoded_test, probs_svc, average='micro')    
    #    svc.fit(X_train_SVD, integer_encoded_train)
    #    probs_SVD = svc.predict(X_test_SVD)
    #    print_AUC(integer_encoded_test, probs_SVD, key)
    #    print_prec_rec(integer_encoded_test, probs_SVD)
    
    logReg_f1 = 0
    if bestClassifier_logReg_config[key] is None:
        print('No logistic regression classifier found for label {}'.format(key))
    else:
        logReg = bestClassifier_logReg_config[key].logReg
        logReg.fit(X_train, integer_encoded_train)
        probs_logReg = logReg.predict_proba(X_test)
        predictions_logReg = convert_probability_list(probs_logReg)      # important because proba returns % for true and false not only true
        pred_bin_logReg = convert_probability_trueFalse(predictions_logReg, 0.5)

        print_AUC(integer_encoded_test, predictions_logReg, key)
        print_prec_rec(integer_encoded_test, predictions_logReg)
        logReg_f1 = f1_score(integer_encoded_test, pred_bin_logReg, average='micro')   
    #    logReg.fit(X_train_SVD, integer_encoded_train)
    #    probs_SVD = svc.predict(X_test_SVD)
    #    print_AUC(integer_encoded_test, probs_SVD, key)
    #    print_prec_rec(integer_encoded_test, probs_SVD)
    
    if svc_f1 > logReg_f1:
        classifierMap[key] = ClassifierContainer(key, svc, tfidf_vectorizer, svd)
    else:
        classifierMap[key] = ClassifierContainer(key, logReg, tfidf_vectorizer, svd)
```

# Predicting unlabeled data

```python
import os
import pandas as pd

script_dir = os.getcwd()
file_path = os.path.join(script_dir, 'Data\\dataDB10000.csv')

data = pd.read_csv(file_path)

relevant_columns = ['doc_text']
newDf = data[relevant_columns].copy()

newDf['cleaned_text'] = clean_text(newDf['doc_text'])
```

```python
for key, classifierContainer in classifierMap.items():   
    tfidf_vectorizer =  classifierContainer.vectorizer
    
    text_cleaned = newDf['cleaned_text']
    text_prepared = tfidf_vectorizer.transform(text_cleaned)
    
    classifier = classifierContainer.classifier   
    prediction = classifier.predict(text_prepared)
    
    newDf[key] = prediction
```
